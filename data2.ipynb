```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('datasets/credits.csv')
credits.head()
```

```python
meta = pd.read_csv('datasets/movies_metadata2.csv')
meta.head()
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')

#If ‘coerce’, then invalid parsing will be set as NaN, if ‘raise’, then invalid parsing will raise an exception, If ‘ignore’, then invalid parsing will return the input
```

```python
# format the date
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in data 1 processed file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = meta.loc[meta.year == 2017, ['genres','id','title','year']]

```

```python

new_meta.head()
```

```python
new_meta.dtypes
```

```python
# converting the 'id' to integer
new_meta['id'] = new_meta['id'].astype(int)
```

```python
# add the new_meta data to the credit data. They all have 'id' so we can merge on that
data = pd.merge(new_meta, credits, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data.head()
```

```python
# we will convert the "genre", "cast" and "crew" column into a list. if we observe it carefully, we will realised that it's a list containing dictionary. 
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))  #the "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

### Genres

```python
# Now we will have to take out the individual genres like adventure, action, sci-fi, etc. Using the following function

def make_genreList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('names') == 'Science Fiction': # Only renaming Science Fiction to Sci Fi to make it shorter, all the other names remains the same
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen)) # then we will join them together and return the values
```

```python
# Now let's apply the function on the genre column

data['genres_list'] = data['genres'].map(lambda x: make_genreList(x))
```

```python
data['genres_list']
```

### Actor 1

```python
# Lets do the same for cast 

def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
# Let's apply it on the cast column for the first actor 
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

### Actor 2

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
# Lets apply it on the cast column for the second actor 
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

### Actor 3

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

### Directors

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
        if dt == []:
            return np.NaN
        else:
            return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
#selecting only the prepared data
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie
```

```python
movie.isna().sum()
```

```python
# renaming 
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
```

```python
# convert all the movie titles to lower case
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
# We will be sing this information later on the tfidvectorizor
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']

movie.head()
```
